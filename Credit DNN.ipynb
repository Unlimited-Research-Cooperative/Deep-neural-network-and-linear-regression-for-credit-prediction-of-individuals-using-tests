Introduction
<br>
<font face="Times New Roman" style="font-size:18px; color:black;">
To solve the root of a psychological problem is much more effective than to treat the symptoms (Lord, 2008). Due to the instant gratification culture (Ray et al., 1986) and strong influence of major pharmaceutical drug companies (Fugh-Berman & Ahari, 2007), the energy and resources of much research and science is allocated to the creation of new treatments of symptoms (Ebeling, 2011). The improperly tended-to economies and cultures of many dominant nations of today can be thought of as the reason for an increase in psychological illness in those nations (Doran & Kinchin, 2017). Financial problems are one of the most common causes of mental illness worldwide (Richardson et al., 2017). To solve global problems related to finance, the usual logical method would include teaching or giving an opportunity to succeed and grow as individuals and communities. Perceived and actual financial literacy separately affect people’s regular personal saving (Nguyen et al., 2017). Both actual and perceived financial literacy appear to influence financial behaviours and that perceived financial literacy may be as important as actual financial literacy (Allgood & Walstad, 2016). Rothwell & Mohammed (2014) developed a questionnaire to assess self-confidence in financial literacy. Respondents with low FICO (Fair Isaac Corporation) scores are far less accurate in their self-assessments (Courchane et al., 2008). Although previous research found that there are correlations between several factors and people’s credit scores, they did not provide a method to predict credit scores. The aim of this thesis is to investigate this hypothesized correlation between credit score and the variables of financial literacy, self-confidence in financial literacy, and IQ, and develop models to predict individual credit scores.
The main purpose of predicting credit scores is to create a method to provide credit to individuals who would otherwise not be able to take loans, or who have not built their credit, or who live in remote locations and do not have access to credit systems. Many more small businesses would have the opportunity to grow, and many more individuals would have access to loans, in which they would be able to study, purchase emergency health products, further themselves in life, improve their standing in the community, and even buy a home. 
A large problem that can be seen in the world today is that middle- and low-income countries are disadvantaged by dominant economies, and, therefore, are dependent on those dominant countries’ economic systems. It is valid to say that this dependency has been very much exacerbated by dominant countries due to their influence on middle- and low-income countries’ regional affairs. Destabilization has occurred in many middle- and low-income countries around the world, which decreases those countries’ economic attractiveness to their own citizens and increases the attractiveness of the dominant countries’ economies. The cognitive dissonance (Festinger, 1962) of dominant countries’ perceived altruism (although being at least partially true due to the surplus of trauma induced empathy)(Greenberg, 2018) in receiving economic migrants by lowering the cost of jobs at the same time may be solved by in fact exporting the strategy and management, not just the labour, of product and service enterprises to middle- and low-income countries. Third wave democracy (Huntington, 1993) in the 1970’s grew democracies around the world from less than 34% of all countries to almost three-quarters of the world. (Karatnycky, 2002). Durkheim’s modernization perspective (1933, 1950) on crime increases due to democratization are but half of the cyclic civilization pathway, since democracy does decrease the criminality of a rising civilization (Weber, 1920; Elias, 1939,1978; Gurr, 1981; Osterberg, 1996; Spierenburg, 1996). Homicide rates on average have doubled in the last 40 years (La Free & Drass, 2002), especially evident in countries where democracy has also grown (Fajnzylber, Lederman, & Loayza, 1998; Diamond, 1999; Mendez et al., 1999; Hraba et al., 1998; Barak, 2000; Backman, 1999; Savelsberg, 1994; Reza Mercy & Krug, 2000).
Through Nietzsche’s idea of ancient Greek world view, it can be said that their ideas were a symptom of decadence (Nietzsche, 1885-1889), and this way of seeing the world is what has given birth to what the majority of dominant powers in the western world publicly states to be the foundation of correct governance and justice. As Plato described, aristocracy degenerates to timocracy, then oligarchy, then, democracy, and then to tyranny, although this could be attributed to western social psychological, evolutionary psychological, and psychological anthropological attributes, since this specific cycle is not evident in the eastern world. Ironically both Plato and Socrates, who can be considered as the founders of western philosophy, agreed on democracy being ultimately made of selfish political and financial figures who merely purchase votes for their and their colleagues’ benefits. Conflict theory (Marx, 1859) supports the logic that the optimistic delusion (Gumplowicz, 1909) of inherent prosperity and safety in the dominant countries of the world is in fact one of the causes for cyclic changes within those countries and their relationship to disadvantaged countries along with increases in  national domestic conflict (Chambliss, 1976; Bohm, 1982). The distribution of global economic resources predominantly to the already dominant countries and individuals aids in the continuation of the lack of equity in the world today (Adams, 1963). According to world system theory (Wallerstein, 1979; Hopkins & Wallerstein, 1981; Chirot & Hall, 1982), distribution is mainly controlled by core countries that have been extracting materials and taking advantage of the low to no cost labour of “peripheral” countries. Anomie (Merton, 1938, Messner & Rosenfeld, 1997), resulting from inequity and social disorganization (Davies, 1962; Smelser, 1962) leads to further cultural decay, a by-product as well as a cause of financial difficulties in individuals. Prospect theory (Kahnemann & Tversky, 1979) can be used as a lens to look upon the history of middle- to low-income countries’ long term economic losses to dominant countries, due to the lack of knowledge the middle- to low-income countries had of game theory (Von Neumann, 1947) and negotiation tactics. Owning only some clothes and shoes, may affect the self-worth of an individual (Covington & Beery, 1976; Covington, 1984), when comparing themselves to others in the world. Owning a home or a business could increase an individual’s self-esteem (James, 1890), which can be considered a social contagion, thus creating positive and healthy cycles of psychological trends in networks connected to each affected individual. 
Britonians that live in owner-occupied dwellings are physically healthier (Fogelman et al., 1989; Blaxter, 1990; Baker & Taylor, 1997), have lower death rates than home-owners (Filakti, 1995), and consult general practitioners less than renters (Carr-Hill, 1996; Benzeval & Judge, 1996). Swedish homeowners are healthier than renters, and consult general practitioners less (Sundquist & Johansson, 1997). Suicide rates are also higher in areas of low homeownership (Ferrada-Noli, 1997), while mental health is higher in areas of high homeownership (Manturuk, 2012). Scots who are homeowners are healthier than renters (Macintyre et al., 1998), and American homeowners have higher functional health (Robert & House, 1996), and even happiness (Rossi & Weber, 1996).
Stable home environments and longer-term neighbourhood relationships increase social capital (Yoo & Lee, 2015). Children of homeowners perform better on both reading and math tests (Essen et al., 1978), regardless of their social class, and children of homeowners are less likely to drop out of high school or have children as teenagers, regardless of race and family income (Green & White, 1997) than children of renters. Children of homeowners grow up to earn greater salaries and are more likely to own a home themselves, regardless of their parents’ income (Boehm & Schottman, 1999). Children of homeowners also have a more stimulating and supportive environment, and less behavioural problems (Haurin et al., 2000). Generally, homeownership is positively related to life satisfaction (Danes & Morris, 1986; Kinsey & Lane, 1983; Lam, 1985; Morris et al., 1976; Varady, 1983; Diaz-Serrano, 2009). Homeowners are more likely than renters to participate in local organizations (Ahlbrandt & Cunninham, 1979; Baum & Kingston, 1984; Cox, 1982; Ditkovsky & van Vliet, 1984; Fischer et al., 1977; Hunter, 1975; Jeffres & Dobos, 1984; Kingston et al., 1984; Lyons & Lowery, 1989; Rohe & Stegman, 1994; Saunders, 1990; Steinberger, 1981; Taub et al., 1977; Wandersman, 1981). Homeownership also improves housing conditions in lower income areas (Galster, 1987; Grigsby, 1963). As homeowners are psychologically healthier than renters, it is logical to assume that helping more people own homes will created more psychologically healthy individuals, families, communities, societies, and relations between nations.

EXPERIMENT 
<br>
Design
<br>
<font face="Times New Roman" style="font-size:18px; color:black;">    
A digital survey was distributed through online websites and chat groups, via Qualtrix, in which participants were asked to input their IQ (a website hyperlink to the Mensa IQ test was provided below the question), credit score (a website hyperlink to the Experian credit score access was provided below the question), answer five questions about how they rate their financial literacy (self-assessment), and answer seven financial literacy questions. I performed analysis using python in Jupyter Notebooks, using correlation matrices and heatmaps, linear regression, multivariable linear regression, and created a neural network. Qualtrix was used as the platform for the survey to be administered and for the survey answers to be compiled and organized into a data frame. Intelligence Quotient (IQ) was included as a variable to explore the influence IQ test scores have on credit scores, due to the fact that IQ includes assessment of general reasoning and creative thinking, both of which can influence an individual’s financial standing. The Mensa IQ test was chosen as the IQ test, as it is a very well-known test with a testing and scoring system generally accepted in scientific research around the world. The financial literacy self-assessment test questions were taken from a previous study on perceived financial capability (Rothman & Khan, 2014). This self-assessment was included to create new scores, which were used to predict the individuals’ credit scores. The self-assessment scores were to be transcribed as 1 (Very Low), 2 (Low), 3 (Moderate), 4 (High), or 5 (Very High), based on their Likert scale. The seven question financial literacy questions were taken from the official FINRA (Financial Industry Regulatory Authority) website, under the financial literacy section (https://www.finra.org/financial_literacy_quiz). The scoring of the five financial literacy answers included one correct answer for each question. Experian was used as the website for participants to be able to be informed what their Experian credit score is, since it is the main credit score company in the United Kingdom and the United States. Correlation matrices were included to show the correlations between individual variables and their correlation to credit score as well. Heatmaps were included to see the correlations more visually between variables. Linear regression was performed to find which variables were most important in exploring and combining with other important variables. Multivariable linear regression was performed to use multiple variables in one linear regression model, due to the high correlation of several variables to credit. An artificial neural network was created, trained, and tested, in order to create a more complex system which is more appropriate for, and can be modified for, big data.

Materials
<font face="Times New Roman" style="font-size:18px; color:black;">    
<br>
The survey in this study consisted of four parts: self-assessment of financial literacy, credit score, IQ score, and a financial literacy test. The survey was created in Qualtrics and performed online, and the data was cleaned, wrangled (transformed and mapped into a usable format), and explored in Excel. The data was then further analysed using Python in Jupyter Notebooks within Anaconda. 
The financial literacy self-assessment section of the survey consisted of five questions from previous research on perceived financial capability (Rothman & Khan, 2014). These questions were: 
	“How would you rate your level of financial knowledge?”
	“How would you rate yourself on keeping track of money?”
	“How would you rate yourself on making ends meet?”
	“How would you rate yourself on ‘shop around to get the best financial product such as loans and insurance rates’”
	“How would you rate yourself on staying informed on financial issues?”
<br>
The financial literacy assessment section consisted of seven questions from the FINRA website (https://www.finra.org/financial_literacy_quiz). These questions were:
	“Suppose you have £100 in a savings account earning 2 percent interest a year. After five years, how much would you have?”
	“Imagine that the interest rate on your savings account is 1 percent a year and inflation is 2 percent a year. After one year, would the money in the account buy more than it does today, exactly the same or less than today?”
	“If interest rates rise, what will typically happen to bond prices? Rise, fall, stay the same, or is there no relationship?”
	“True or false: A 15-year mortgage typically requires higher monthly payments than a 30-year mortgage but the total interest over the life of the loan will be less.” 
	“True or false: Buying a single company's stock usually provides a safer return than a stock mutual fund.”
	“Suppose you owe £1000 on a loan and the interest rate you are charged is 20% per year compounded annually. If you didn't pay anything off, at this interest rate, how many years would it take for the amount you owe to double?”
	“Which of the following indicates the highest probability of getting a particular disease?”
<br>
The Experian credit score of participants could be found by them on the Experian website (https://www.experian.com), and the Mensa IQ test could be found on the Mensa website (https://www.mensa.org/public/mensa-iq-challenge).  The Python libraries used for  analysis and models were Pandas, NumPy, Seaborn, MatPlotLib, PyLab, SKLearn, math, Bezier, ForceAtlas2, jupyterthemes, ANN_visualizer, TensorFlow, and Keras.

Statistical Analysis Procedure
<font face="Times New Roman" style="font-size:18px; color:black;">    
<br>
The survey answers in Qualtrics were exported in xlsx format, then uploaded into Microsoft Excel. Data was cleaned and wrangled in several steps. Incomplete participant survey results and out of bounds credit scores were removed. Whole numbers, “1” through “5”, were assigned to all five options of the Likert scales “Very Low,” “Low,” “Moderate,” “High,” “Very High” in the financial literacy self-assessment data columns. “0” was assigned to all incorrect answers and “1” to all correct answers in the financial literacy assessment data columns. A data column was created with the summed totals of each participant’s financial literacy self-assessment and another data column was created with the summed totals of each participants financial literacy assessment. The totals were divided by the maximum possible totals (thus forming a percent score) and two additional data columns were created containing these scores for each participant. The financial literacy self-assessment score column was then subtracted from the financial literacy score column, to give an adjusted score, and the financial literacy score column was then divided by the financial literacy self-assessment score column, to give a secondary adjusted score. The financial literacy self-assessment score column was then divided by the financial literacy score column, to give a third adjusted score. The cleaned data was then uploaded into Jupyter Notebooks via Python, and the data was split into “X” and “Y” values. All data columns except the credit scores column were assigned to “X” and the credit scores were assigned to “y”. The data was then split into training and testing data, with a test size of 0.33 and a random state of 100.  
The variables “A1” (How would you rate your level of financial knowledge), “A2” (How would you rate yourself on keeping track of money?), “A3” (How would you rate yourself on making ends meet?), “A4” (How would you rate yourself on “shop around to get the best financial product such as loans and insurance rates”?), “A5” (How would you rate yourself on staying informed on financial issues?), comprise of the data from the participants’ answers to questions 1, 2, 3, 4, and 5, in the financial literacy self-assessment. The variable “IQ” is the IQ score, and the variables “B1” (Suppose you have £100 in a savings account earning 2 percent interest a year. After five years, how much would you have?), “B2” (Imagine that the interest rate on your savings account is 1 percent a year and inflation is 2 percent a year. After one year, would the money in the account buy more than it does today, exactly the same or less than today?), “B3” (If interest rates rise, what will typically happen to bond prices? Rise, fall, stay the same, or is there no relationship?), “B4” (True or false: A 15-year mortgage typically requires higher monthly payments than a 30-year mortgage but the total interest over the life of the loan will be less.), “B5” (True or false: Buying a single company's stock usually provides a safer return than a stock mutual fund.), “B6” (Suppose you owe £1000 on a loan and the interest rate you are charged is 20% per year compounded annually. If you didn't pay anything off, at this interest rate, how many years would it take for the amount you owe to double?), and “B7” (Which of the following indicates the highest probability of getting a particular disease?) comprise of the data from questions 1, 2, 3, 4, 5, 6, and 7, from the participants’ answers to questions in the financial literacy assessment. Since complexity in models can improve accuracy, the C_(i )variables were calculated by creating scores from the financial literacy test and the financial literacy self-assessment test, and combining variables with use of increased differentiation. “C1” is the score of the financial literacy self-assessment, and “C2” is the score of the financial literacy assessment. “C1” was calculated by dividing each participant’s total score from the financial literacy self-assessment by 25, since the maximum score for each answer to assessment questions was five, and there were five questions. “C2” was calculated by dividing the total score from the financial literacy assessment by seven, since the maximum score for each answer was one, and there were seven questions.
C1=(A1+A2+A3+A4+A5)/25
C2=(B1+B2+B3+B4+B5+B6+B7)/7
“C3” is the financial literacy self-assessment score subtracted from the financial literacy assessment score and “C4” is the financial literacy assessment score divided by the financial literacy self-assessment score. The reason for the use of subtraction and division was to combine the variables to create a new variable, due to subtraction as a method of differentiating scores, and division as a method of expanding that differentiation. 
C3=C2-C1
C4=C2/C1
<br>
Further calculations were performed and assigned to “C5,” “C6,” “C7,” “C8,” and “C9,” to create more complexity of variables. “C5” is the financial literacy self-assessment score divided by the financial literacy self-assessment score. The reasoning behind this formula was to create an alternate combined score of “C1” and “C2.”
C5=C1/C2
“C6” is the sum of “B2” and “B6”, divided by two, divided by the sum of “A2” and “A5”, divided by two. “C7” is the sum of “B2” and “B6”, divided by two, subtracted by the sum of “A2” and “A5”, divided by two. The reasoning behind the “C6” formula was to only use the questions with highest correlation to credit scores from the financial literacy self-assessment test and financial literacy test, division by two was performed on the pairs of test answer variables since they are pairs, to find the mean of each new pair score. Division was further performed to create model complexity. “C7” was calculated using subtraction instead of division, to create alternative model complexity.
C6=(B2+B6)/2÷(A2+A5)/2
C7=(B2+B6)/2-(A2+A5)/2
“C8” is the sum of “B2” and “B6”, divided by two, divided by the sum of “A2” and “A5”, divided by ten. “C9” is the sum of “B2” and “B6”, divided by two, subtracted by the sum of “A2” and “A5”, divided by ten. The reasoning behind the “C8” formula of dividing the question variable pair from the financial literacy self-assessment test by ten instead of two was to create new scores from the two pairs of variables. The variable pair (B2 and B6) from the financial literacy test had a maximum possible score of two, and the variable pair (A2 and A5) from the financial literacy self-assessment test had a maximum score of 10.
C8=(B2+B6)/2÷(A2+A5)/10
C9=(B2+B6)/2-(A2+A5)/10
Microsoft Excel was used to find descriptive statistics for the cleaned dataset. Questions “A2”, “A5”, “B2”, and “B6” were chosen due to their correlations to credit being the highest in their respective tests. Multiplicative inverse was used to increase the complexity of values. The correlations between credit and all other variables were then calculated (see Figure 5) and visualised on a heatmap (see Figure 6) in Seaborn with Python. This was used to show how all the variables are related to each other, with ease of human processed pattern recognition and component identification.

Discussion
The multivariable linear regression model shows a meaningful correlation between credit scores and the independent variables, which can be an option for creditors to predict credit scores in individuals without credit scores, or even give an alternatively more accurate credit scoring system to individuals who already have credit scores. The financial literacy self-assessment was vital to the accuracy of the models. The credit prediction artificial neural network did not perform as well as the credit prediction linear regression, most likely due to a small dataset to train on. To increase the accuracy of the artificial neural network, more data could be gathered from the same variables included in the dataset, on additional participants. Layers and functions could be added or changed, as well as node pruning performed. The numbers in each of the questions could be randomized to avoid cheating on the test, since almost all of the questions on the financial literacy test include numbers or have answers that include numbers. The question with the highest correlation to credit scores was “C3”, “If interest rates rise, what will typically happen to bond prices? Rise, fall, stay the same, or is there no relationship?” and a possible problem of test takers cheating on the test could be solved by rewriting this question as, for example, “If interests rates change from 7% to 7.25%, is it more probable that bond prices would go from 1240 to 1290, from 1240 to 1215, or stay the same?”
<br>
This artificial neural network could be a more sophisticated option for creditors to predict credit scores of individuals with low credit scores or without credit scores. Possible variables to be added could include individuals’ social media actions (for example, abusive speech, scamming), app payments, public disturbances, traffic rule violations, community service, tardiness to hospital/government/work/school appointments, personality traits, genetics, and biometrics from current technologies such as the five-second intermittent infrared face capture that is performed by smartphones while looking at them, eye tracking and pupil dilation tracking in virtual reality headsets, and phone/computer app usage/activity statistics. Genetics can be used due to polygenic scores being able to be used to predict adult economic outcomes even after accounting for educational attainments (Belsky et al., 2016). Sesame Credit and the Social Credit System in China are examples of complex algorithmic credit scoring and social scoring and have had much western media publicity in recent years. Although there are many who oppose mass data collection of individuals due to privacy and freedom concerns, it is realistic to assume that the majority of humans, whether in the west or east, do not live in a reality where they as individuals are actually governed by bodies who obligate themselves to give their governed humans freedoms of speech, movement, privacy of information/behaviour, or the usual topics of claims that are made against big data mining of individuals’ information/behaviour. Companies such as Neo Finance, FriendlyScore, Affirm and Kabbage already use social media for their credit scoring of individuals. Future research to further improve accuracy of the credit prediction artificial neural network could include exploring mathematics testing and adding more dependent variables such as history of individuals’ defaults and bankruptcy. The addition of more dependent variables could be used to replace the current credit scoring methods completely.

Banks, credit unions and P2P lenders could use the Credit Score Prediction Model to help  dysfunctional families, which are linked to mental disorders (Crittenden et al., 1991; Allen, 2010), around the world, by being given the opportunity to own a house through a mortgage, which could vastly improve the psychological health of their future generations.
